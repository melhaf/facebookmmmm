### 4)  Correlation

```python
corr = mm_dataset.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(mm_dataset.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(mm_dataset.columns)
ax.set_yticklabels(mm_dataset.columns)
plt.show()
```

*Correlation matrix*

The darkest red means there is a perfect positive correlation, while the darkest blue means there is a perfect negative correlation.

a) The two most strongly correlated is: $v_4$ with $v_5$, and $v_2$ with $v_1$.

b) The least correlated variables is: $v_4$ with $v_3$.
